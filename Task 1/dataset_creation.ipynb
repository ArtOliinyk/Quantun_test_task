```python
import random

# List of famous mountain names
mountain_names = [
    "Mount Everest", "K2", "Kangchenjunga", "Lhotse", "Makalu", 
    "Cho Oyu", "Dhaulagiri", "Manaslu", "Annapurna", "Mont Blanc", 
    "Mount Kilimanjaro", "Mount Elbrus", "Denali", "Aconcagua", "Mount Fuji"
]

# Sentence templates
sentence_templates = [
    "I plan to climb {} next summer.",
    "Have you ever been to {}?",
    "{} is one of the tallest peaks in the world.",
    "The view from the top of {} is breathtaking.",
    "We are planning an expedition to {} next year.",
    "{} has always been on my bucket list.",
    "The summit of {} is covered in snow all year round.",
    "Many mountaineers attempt to conquer {} every year.",
    "{} is known for its incredible views.",
    "Climbers often talk about the challenges of scaling {}.",
    "{} is famous for its extreme weather conditions.",
    "We visited the base camp of {}.",
    "Do you know the history of expeditions to {}?",
    "Reaching the summit of {} is a dream for many climbers.",
    "Hiking around {} is an unforgettable experience.",
    "The glaciers of {} are stunning.",
    "Scaling {} is considered a major achievement.",
    "{} is part of the Seven Summits challenge.",
    "I hope to reach the top of {} one day.",
    "Several expeditions to {} have been successful this year.",
    "{} is located in a remote area with harsh conditions.",
    "The weather near {} can change rapidly.",
    "{} is often attempted by experienced mountaineers.",
    "During the summer, we will try to summit {}.",
    "I read an article about a recent expedition to {}.",
    "{} offers stunning panoramic views from its summit.",
    "Many climbers train for years to attempt {}.",
    "There are multiple routes to the summit of {}.",
    "The altitude of {} poses a significant challenge.",
    "{} is a popular destination for adventurous travelers.",
    "Standing on top of {} feels like being on top of the world.",
    "The base camp of {} attracts adventurers from all over the world.",
    "Photographers love capturing the beauty of {}.",
    "{} is a difficult peak to climb due to its technical demands.",
    "We are preparing for a climb to {}.",
    "{} is a less crowded mountain compared to some others.",
    "The local guides know the best routes to summit {}.",
    "Several teams have attempted to climb {} this season.",
    "The landscape around {} is truly breathtaking.",
    "{} is part of the Himalayas and stands as a symbol of beauty.",
    "Climbers usually spend weeks acclimatizing before attempting {}.",
    "There are strict environmental regulations around {}.",
    "{} stands as one of the most iconic peaks in the world.",
    "In the winter, climbing {} becomes even more challenging.",
    "The ascent of {} is both mentally and physically demanding.",
    "{} is the highest mountain in its region.",
    "Mountaineers often face severe weather conditions on {}.",
    "{} is known for its long and dangerous ascent.",
    "The locals have a rich history of guiding climbers to the top of {}.",
    "I met a group who had just returned from their climb of {}.",
    "The wildlife near {} is as stunning as the mountain itself.",
    "After months of training, Iâ€™m finally ready to climb {}."
]

# Function to generate 200 synthetic sentences with random mountain names
def generate_synthetic_sentences(num_sentences=200):
    sentences = []
    for _ in range(num_sentences):
        template = random.choice(sentence_templates)
        mountain = random.choice(mountain_names)
        sentence = template.format(mountain)
        sentences.append(sentence)
    return sentences

# Generate 200 synthetic sentences
synthetic_sentences = generate_synthetic_sentences(200)

# Display a few sentences
for sentence in synthetic_sentences[:10]:
    print(sentence)
```

```python
# Function to label mountain names in the sentences
def label_sentences(sentences, mountain_names):
    labeled_data = []
    
    for sentence in sentences:
        tokens = sentence.split()  # Split sentence into words
        labels = []
        for token in tokens:
            # Check if the token matches any mountain name or part of it
            found = False
            for mountain in mountain_names:
                if token in mountain.split():
                    if token == mountain.split()[0]:
                        labels.append('B-Mountain')  # Beginning of mountain name
                    else:
                        labels.append('I-Mountain')  # Inside mountain name
                    found = True
                    break
            if not found:
                labels.append('O')  # Other tokens not part of an entity
        labeled_data.append(labels)
    
    return labeled_data
```

```python
import pandas as pd

# List of sentences (synthetic sentences generated previously)
synthetic_sentences = generate_synthetic_sentences(5000)

# Corresponding NER labels (BIO format for the sentences)
labeled_sentences = label_sentences(synthetic_sentences, mountain_names)

# Create a pandas DataFrame with two columns: 'text' and 'labels'
data = {
    'text': synthetic_sentences,
    'labels': [' '.join(labels) for labels in labeled_sentences]  # Join the labels for each sentence
}

# Create the DataFrame
df = pd.DataFrame(data)

# Display the first few rows of the DataFrame
df.head()
```
